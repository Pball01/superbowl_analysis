```python
# Setting variables to be used below
maxTweets = 500

# Creating list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:Budwiser').get_items()):
    if i>maxTweets:
        break
    tweets_list1.append([tweet.url, tweet.date, tweet.id, tweet.content.replace('\n', ' '), 
                        tweet.retweetCount, tweet.replyCount,tweet.likeCount,tweet.quoteCount, 
                        tweet.username,tweet.user.displayname, tweet.user.followersCount, tweet.user.friendsCount, 
                        tweet.lang, tweet.mentionedUsers, tweet.retweetedTweet, tweet.])

# Creating a dataframe from the tweets list above
tweets_df1 = pd.DataFrame(tweets_list1, columns=['url','Datetime', 'Tweet Id', 'Text', 'Retweet Count', 'Reply Count', 'Like Count',
                            'Quote Count', 'Username', 'Display Name','Followers Count', 'Friends Count', 'Language', 'MentionedUsers',
                             'RetweetId', 'Date'])

# Display first 5 entries from dataframe
tweets_df1.head()
```
